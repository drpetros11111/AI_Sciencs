```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.feature import hog
from skimage import data, exposure
```

```python
from skimage.feature import hog
from skimage import data, exposure
import matplotlib.pyplot as plt

# Load the example image
image = data.astronaut()

# Compute HOG features and the HOG image
fd, hog_image = hog(image, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True, channel_axis=-1)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)

# Display the original image
ax1.axis('off')
ax1.imshow(image, cmap=plt.cm.gray)
ax1.set_title('Input image')

# Rescale histogram for better display
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))

# Display the HOG image
ax2.axis('off')
ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray)
ax2

```

# HOG -Histogram of Oriented Gradients

It's a feature descriptor used in computer vision and image processing for object detection.

HOG captures the gradient structure or the edge information in an image, which is particularly useful for detecting objects and shapes.

 HOG captures the gradient structure or the edge information in an image, which is particularly useful for detecting objects and shapes.

 Here’s how it works:

##Gradient Computation

 HOG calculates the gradient of the image intensity in both horizontal and vertical directions. This helps to capture the edges and their orientations.

##Orientation Binning

The image is divided into small connected regions called cells, and for each cell, a histogram of gradient directions (or orientations) is compiled.

This histogram represents the distribution of gradient directions within the cell.

##Descriptor Blocks

To improve robustness to lighting, shadowing, and other variations, these cells are grouped into larger spatial regions called blocks.

The HOG descriptor is then normalized within these blocks, making it less sensitive to changes in illumination or contrast.

##Feature Vector

The normalized histograms from all the blocks in the image are concatenated to form a single feature vector representing the image.

HOG calculates the gradient of the image intensity in both horizontal and vertical directions.

The process of calculating the gradient of the image intensity in both horizontal and vertical directions is a crucial step in the Histogram of Oriented Gradients (HOG) feature extraction process. Here's an explanation of what this means and why it's done:

##Image Gradient
An image gradient is a directional change in the intensity or color in an image.

In the context of HOG, we are interested in capturing the edges and shapes within the image, which can be detected by calculating the gradients.

##How Gradients Are Calculated

##Horizontal Gradient (Gx)

This represents the change in intensity from left to right in the image. It is calculated by convolving the image with a kernel that highlights horizontal changes. A common choice is the Sobel operator, which uses the following kernel:

image.png



This kernel emphasizes the differences in pixel values along the horizontal direction, effectively highlighting vertical edges.

##Vertical Gradient (Gy):

This represents the change in intensity from top to bottom in the image. It is calculated by convolving the image with a kernel that highlights vertical changes. The Sobel operator for the vertical gradient is:

image.png



This kernel emphasizes the differences in pixel values along the vertical direction, effectively highlighting horizontal edges.

#Why Both Directions?

image.png


By calculating the gradient in both horizontal and vertical directions, we can capture the intensity changes in all possible directions.

This is important because edges and features in an image can have any orientation. The combined information from both gradients allows us to compute the magnitude and direction (orientation) of the gradient at each pixel:


##Example
Consider a simple 3x3 grayscale image:

    10  10  10
    10  20  10
    10  10  10

The horizontal gradient at the center pixel (20) would be calculated using the horizontal Sobel operator. The surrounding horizontal changes are:


    -1*10 + 0*10 + 1*10
    -2*10 + 0*20 + 2*10
    -1*10 + 0*10 + 1*10

This results in $$Gx = 0$$.

The vertical gradient at the same center pixel would be calculated using the vertical Sobel operator. The surrounding vertical changes are:


    -1*10 + -2*10 + -1*10
     0*10 +  0*20 +  0*10
     1*10 +  2*10 +  1*10

This results in $$Gy = 0$$.

In this example, the gradients are zero because the central pixel (20) is surrounded by pixels of equal intensity (10), indicating no edge. For more complex images, these gradients help identify edges and their orientations.

##Summary
Calculating the gradient of the image intensity in both horizontal and vertical directions allows us to detect edges and their orientations in an image.

These gradients form the basis for constructing the histograms of oriented gradients, which are used to describe the image's features in a robust and informative way for tasks like object detection and recognition.

The calculated gradient directions are continuous values (angles), typically ranging from 0 to 360 degrees (or 0 to 2π radians).

To build the HOG descriptor, these continuous directions are quantized into a fixed number of bins. This process is known as orientation binning.
orientations=8:

Setting orientations=8 means that the continuous gradient directions will be quantized into 8 bins.

Each bin represents an angular range of 45 degrees (360 degrees / 8 bins). For instance:

Bin 1: 0 - 45 degrees

Bin 2: 45 - 90 degrees

Bin 3: 90 - 135 degrees

Bin 4: 135 - 180 degrees

Bin 5: 180 - 225 degrees

Bin 6: 225 - 270 degrees

Bin 7: 270 - 315 degrees

Bin 8: 315 - 360 degrees

---

##Create Histograms

For each cell in the image (e.g., a 16x16 pixel region, as specified by pixels_per_cell), a histogram of gradient orientations is created.
Each gradient within the cell contributes to the histogram according to its magnitude and falls into one of the 8 bins based on its direction.

===
##Feature Vector

The histograms from all cells are concatenated to form the final HOG feature vector for the image.

This feature vector is then used in machine learning models for tasks such as object detection or image classification.

---
##Summary
By setting orientations=8, you are telling the HOG algorithm to consider 8 different gradient directions (or orientations) when building the histograms.

This determines the level of detail captured about the gradient directions in the image, influencing the granularity and effectiveness of the feature representation for distinguishing between different patterns in the image.
