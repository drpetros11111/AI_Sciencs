#Perceptron


```python
def prediction(x, W, b):
    return np.dot(x, W) + b >= 0

def perceptron(X, y, W, b, lr):
    for i in range(len(X)):
        y_hat = prediction(X[i], W, b)

        if y[i] - y_hat == 1:
            W += X[i] * lr
        elif y[i] - y_hat == -1:
            W -= X[i] * lr
            b -= lr

    return W, b

def trainPerceptron(X, y, lr, n_epochs):
    W = np.random.rand(X.shape[1])
    b = 2
    lines = []

    for i in range(n_epochs):
        lines.append((-W[0] / W[1], -b / W[1]))
        W, b = perceptron(X, y, W, b, lr)

    return lines
```

##**Define the Prediction Function**

**def prediction(x, W, b):**

This function is defined to predict the class label for a given input sample x using weights W and bias b.

**np.dot(x, W) + b:**

**np.dot(x, W) **

Computes the dot product between the input sample x and the weight vector W.

**+ b**

Adds the bias term b to the result of the dot product.

**>= 0**

Checks if the result of the dot product plus the bias is greater than or equal to 0.

This returns a boolean value: True (1) if the condition is met, and False (0) otherwise.

This effectively classifies the sample as 1 if the result is non-negative and as 0 otherwise.



---



##**Define the Perceptron Function**

**def perceptron(X, y, W, b, lr):**


This function implements the perceptron learning algorithm to update the weights W and bias b based on the training data X and y.

**for i in range(len(X)):**

Iterates over each sample in the training data X.

**y_hat = prediction(X[i], W, b):**

Uses the prediction function to compute the predicted label y_hat for the i-th sample.

**if y[i] - y_hat == 1:**

If the true label y[i] is 1 and the predicted label y_hat is 0, the difference y[i] - y_hat equals 1. ***This indicates a false negative.***

**W += X[i] * lr:**

Updates the weights by adding the scaled input sample X[i] to W.
Explanation: This update moves the decision boundary to correctly classify the sample as 1.

**elif y[i] - y_hat == -1:**

If the true label y[i] is 0 and the predicted label y_hat is 1, the difference y[i] - y_hat equals -1. This indicates a false positive.

**W -= X[i] * lr:**

Updates the weights by subtracting the scaled input sample X[i] from W.
b -= lr: Decreases the bias by the learning rate lr.

**Explanation**

This update moves the decision boundary to correctly classify the sample as 0.
return W, b:

Returns the updated weights and bias after one pass through the training data.



---
##**Function TrainPerceptron **bold text**

**def trainPerceptron(X, y, lr, n_epochs):**

This function trains the perceptron model over multiple epochs.

**W = np.random.rand(X.shape[1])**:

Initializes the weight vector W with random values. The length of W is equal to the number of features in the input data X.

**b = 2:**

Initializes the bias b with a value of 2.

**lines = []**

Initializes an empty list lines to store the slope and intercept of the decision boundary after each epoch.

**for i in range(n_epochs):**

Iterates over the number of epochs specified by n_epochs.

**lines.append((-W[0] / W[1], -b / W[1]))**

Computes the slope and intercept of the decision boundary based on the current weights and bias and appends them to the lines list.

**W, b = perceptron(X, y, W, b, lr):**

Calls the perceptron function to update the weights W and bias b using the training data X and y.

**return lines:**

Returns the list of slope and intercept pairs representing the decision boundary after each epoch.



---


**Summary**

**prediction**

Computes the predicted class label for a given input sample.

**perceptron**

Updates the weights and bias based on the training data using the perceptron learning rule.

**trainPerceptron**

Trains the perceptron model over multiple epochs and records the decision boundary parameters after each epoch.



```python
res = trainPerceptron(X.T, y, lr=0.01, n_epochs=35)
```

```python
# Define the abline function outside the loop
def abline(slope, intercept):
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')

# Assuming res is a list of [slope, intercept] pairs
res = [
    [1, 0],  # Example values
    [-1, 2]  # Example values
]

# Plotting loop
for i in range(len(res)):
    plt.scatter(X[0][np.where(y == 1)], X[1][np.where(y == 1)], color='blue', label='Class 1' if i == 0 else "")
    plt.scatter(X[0][np.where(y == 0)], X[1][np.where(y == 0)], color='red', label='Class 0' if i == 0 else "")
    abline(res[i][0], res[i][1])
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(f'Plot with Line {i + 1}')
    if i == 0:
        plt.legend()
    plt.show()
```
