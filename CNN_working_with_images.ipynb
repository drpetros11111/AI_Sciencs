# 1. Converting the Image to Grayscale

    imGray = 0.333*R + 0.333*G + 0.333*B


---


## R, G, B

These are 2D arrays representing the Red, Green, and Blue channels of the image respectively.

Each array has the same shape as the original image, (height, width).


---


    0.333*R

This multiplies each value in the Red channel by 0.333.

The coefficient 0.333 is approximately one-third, which means we are taking an equal portion of the Red channel.


---



    0.333*G

This multiplies each value in the Green channel by 0.333.

----
    0.333*B

This multiplies each value in the Blue channel by 0.333.

----
##Addition

Adding these three results combines the contributions of each color channel into a single 2D array.

Each pixel's intensity in imGray is calculated as the average of its Red, Green, and Blue values.

----
##Result

imGray is a 2D array representing the grayscale image where each pixel intensity is the average of the RGB values.

This is a simple method to convert an RGB image to grayscale.

----
----

#2. Displaying the Grayscale Image

    plt.imshow(imGray, cmap='gray')

##plt.imshow

##Result

The plt.imshow function will display the grayscale image using a grayscale colormap.

Each pixel's intensity in imGray will be represented in shades of gray.

---

#Summary

imGray: A grayscale version of the original image obtained by averaging the Red, Green, and Blue channels.

plt.imshow(imGray, cmap='gray'): Displays the grayscale image using Matplotlib.

This process effectively converts the RGB image into a single-channel grayscale image by averaging the color intensities.

# Blurring, Edge, Sharpening

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy import signal
```

```python
im = plt.imread(r'/content/bird.png')[:,:,:3]
```

```python
imGray = cv2.cvtColor(im,cv2.COLOR_RGB2GRAY)
```

# 1. Using CV2 Library
    cv2.cvtColor

    cv2

This is the OpenCV library, a popular library for computer vision tasks.

You need to have OpenCV installed (pip install opencv-python) to use this function.

    cv2.cvtColor

This function is used to convert an image from one color space to another.


---


#2. Parameters

    im

This is the input image in RGB format. In OpenCV, images are often represented in BGR format by default, but here we assume it is in RGB format.

    cv2.COLOR_RGB2GRAY

This is a flag that tells cv2.cvtColor to convert the image from RGB (Red, Green, Blue) to Grayscale.

OpenCV uses this flag to know the conversion mode.


---


#3. Conversion Process

The cv2.cvtColor function converts the input image from the RGB color space to grayscale.

In grayscale images, each pixel value represents the intensity of light, with no color information.

The grayscale conversion is typically done using a weighted sum of the RGB channels. For standard conversion, the formula used is often:

$$Gray
=
0.299


+
0.587


+
0.114

$$


Here, the weights reflect human perception of different colors, which is why the Green channel has the highest weight.

----
#4. Output

    imGray

This is the output grayscale image, which is a 2D array where each value represents the intensity of a pixel.

It is derived from the original RGB image by removing the color information and only retaining the intensity.

----
#Summary
cv2.cvtColor(im, cv2.COLOR_RGB2GRAY):

Converts the input RGB image im into a grayscale image imGray.

imGray:

2A 2D array where each pixel value represents the grayscale intensity of the corresponding pixel in the original RGB image.

This conversion is commonly used in various image processing tasks where color information is not necessary, such as edge detection or image thresholding.

```python
smoothingMask = np.ones((15,15))/225
```

```python
imBlurred = signal.convolve2d(imGray,smoothingMask,
                             boundary='symm',mode='same')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(10,15))
ax1.imshow(imGray,cmap='gray')
ax2.imshow(imBlurred,cmap='gray')
```

```python
xMask = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
yMask = xMask.T.copy()
```

```python
fx = signal.convolve2d(imGray,xMask,boundary='symm',mode='same')
fy = signal.convolve2d(imGray,yMask,boundary='symm',mode='same')
```

```python
fx.shape
```

```python
imGray.shape
```

```python
Gm = (fx**2 + fy**2)**0.5
```

```python
plt.imshow(Gm,cmap = 'gray')
```

```python
th = Gm.max()-4*Gm.std()
```

```python
th
```

```python
E = Gm>th
plt.imshow(Gm>0.1,cmap = 'gray')
```

```python
imSharped = imBlurred + 0.5*Gm
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(10,15))
ax1.imshow(imBlurred,cmap='gray')
ax2.imshow(imSharped,cmap = 'gray')
```
